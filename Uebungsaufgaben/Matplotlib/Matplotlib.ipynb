# Übungen zur Matplotlib

Für jede der folgenden Übungsaufgaben gibt es ein Bild, welches nachgebaut werden soll.

Zuerst müssen jedoch die Daten geladen werden:

````python
import json
import matplotlib.pyplot as plt

with open("barplot.dat", "r") as data:
    barplot = json.load(data)
with open("histogram.dat", "r") as data:
    histplot = json.load(data)
with open("square.dat", "r") as data:
    xdata, ysquare = json.load(data).values()
with open("powers.dat", "r") as data:
    ypower = json.load(data)["y"]
with open("exponential.dat", "r") as data:
    yexponential = json.load(data)["y"]
````

## 1a) Einfache Plots

Tipp: Geben Sie jeder Zeichnung einen Namen und Größe mit
```Python
plt.figure(num="Name der Zeichnung", figsize=[7, 7], dpi=100)
```

````python
# YOUR CODE HERE
plt.plot(xdata, ysquare, label='ysquare')
plt.legend()
plt.ylabel('y_data')
plt.xlabel('x_data')
plt.show()
````

## 1b) Mehrere Daten in einer Zeichnung

````python
plt.plot(xdata, ysquare, label='ysquare', marker='X', LineStyle='None', c='red')
plt.plot(xdata, ypower, label='power', c='darkcyan', linewidth=5)
plt.legend()
plt.ylabel('y_data')
plt.xlabel('x_data')
plt.show()
````

## 1c) Exponentielle Daten
Für Daten mit exponentiellem Charakter ist eine Darstellung mit Logarithmischer Achse hilfreich

````python
# YOUR CODE HEREplt.plot(xdata, ysquare, label='ysquare', marker='X', LineStyle='None', c='red')
plt.plot(xdata, yexponential, label='yexponential', c='red', marker='X', LineStyle='dashed')
plt.legend()
plt.ylabel('y_data')
plt.xlabel('x_data')
plt.yscale('log')
plt.xscale('log')
plt.show()
````
